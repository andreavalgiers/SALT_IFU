```python
import numpy as np
import astropy
import scipy
import matplotlib.pyplot as plt
import pyfits
import itertools
from astropy.stats import sigma_clip
import pylab as pl
import os
import matplotlib.animation as animation
import sys
from scipy import stats
from scipy import interpolate
from scipy.interpolate import UnivariateSpline
import math
import pandas
```

```python
#### SALT TELESCOPE PARAMTERS

collFL = 647.09 # Collimator Focal Length [mm] 
camFL = 220.0 # Camera Focal Length [mm] 
platescale=18 # Platescale [micron pix^-1]
focalscale = 226.178 # focal scale [microns arcsecs^-1]
slitwidth= 0.9 # Slit Width [arcsec]
groove_sep=1./950. # Groove Separation [mm] 
primD=11 # Primary Mirror Diameter [m]
secD=2 #Secondary Mirror Diameter [m]
collectArea=91.89 #Collecting Area [m^2]
eff_RSS=0.2 # Efficiency of RSS instrument
eff_fib=0.78 # Efficiency of fiber
resem=4.233 # Resolution Element [pix]
totpix=2048 #Number of pixels in detector
slitpix=2031.84 #Number of pixels along slit length
pixArea=0.000324 # Area of Pixel mm^2
RN = 9 # Read Noise
npix = 17.92 # 
gain = 2.5 #
dark = 0.002 #
```

```python
#### Calculation of Spectrograph

### Grating Equation= Central Wavelength
alpha=arange(20,55,5)
cent_wave=2*groove_sep*np.sin(alpha*np.pi/180.)*10**6 # [nm]

## Linear Dispersion
lin_disp=0.0861*np.cos(alpha*np.pi/180.) #[nm pix^-1]

## Fiber Size
fibersize=np.array([100.,150.,200.,300.,400.,500.,600.,800.])/focalscale # Diamter of fiber [arcsec]
```

```python
### Sky Continuum
skywave = np.array([900, 950, 1000, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 
   1400, 1450, 1500, 1550, 1600, 1650, 1700, 1750, 1800, 1850, 1900]);
skyflux = np.array([1.315709, 0.3790091, 1.11682, 1.045974, 0.9701224, 
   0.9064792, 0.9064792, 0.9064792, 0.9064792, 0.9064792, 0.9064792, 
   0.9064792, 0.9064792, 0.9064792, 0.9064792, 0.9064792, 0.9064792, 
   0.9064792, 0.9064792, 0.9064792, 0.9064792]);

## Interpolation of skywave and skyflux
f = interpolate.interp1d(skywave, skyflux,bounds_error=False, fill_value=1.315709)

### Sky Continuum At The Central Wavelengths
### What we actually measureed (after atmospheric tranparency and efficiency)

trans=np.array([0.948654,0.948654,0.99985,0.976124,0.356107,0.842272,0.982687]) # Atmospheric Transmission at Mauna Kea as a function
                                                                       # alpha
rss_eff=np.array([0.103906905,0.103906905,0.360919446,0.462250431,0.421808381,0.438769116,0.38712636]) # RSS efficiency at various alpha angles


skyflux_det=[]
for i in range(0,len(cent_wave)):
    
    skyflux_cent=f(cent_wave[i])*collectArea*np.pi*((fibersize/2)**2)*lin_disp[i]*trans[i]*rss_eff[i]*eff_fib # photons s^-1 pix
    skyflux_det.append(skyflux_cent)

skyflux_det=np.array(skyflux_det)# The detected sky intensity as a function of angle and fiber size
                                  # [photons s^-1 pix^1]
    
```

```python
print np.transpose(skyflux_det)[1]*gain
```

```python
### Sky-Limited vs. Read-Noise
#### Sky-Limited: Sqrt(skyflux_det*exposure_time)
#### Read-Noise Limited: 3*RN



fib_sky_alpha= np.transpose(skyflux_det) # Gives the sky flux for each alpha for each fiber size

exp_time=[1.,60.,100.,300.,600.]
RN_lim=np.array( [3*RN]*(len(alpha)-1))

sky_lim=[]

for k in range(0,len(exp_time)):
    arr=[]
    for i in range(len(fib_sky_alpha)):
        arr.append(np.sqrt(fib_sky_alpha[i]*gain*exp_time[k])) # Units of electrons per pix
        #arr.append(fib_sky_alpha[i]*exp_time[k])
    sky_lim.append(arr)

```

```python
### The detected sky flux as a function of alpha for each fiber size. Units of photons/pix

## At exposure time t=1 sec
fib100_0=sky_lim[0][0][1:]
fib150_0=sky_lim[0][1][1:]
fib200_0=sky_lim[0][2][1:]
fib300_0=sky_lim[0][3][1:]
fib400_0=sky_lim[0][4][1:]
fib500_0=sky_lim[0][5][1:]
fib600_0=sky_lim[0][6][1:]
fib800_0=sky_lim[0][7][1:]

## At exposure time t=60 sec
fib100_1=sky_lim[1][0][1:]
fib150_1=sky_lim[1][1][1:]
fib200_1=sky_lim[1][2][1:]
fib300_1=sky_lim[1][3][1:]
fib400_1=sky_lim[1][4][1:]
fib500_1=sky_lim[1][5][1:]
fib600_1=sky_lim[1][6][1:]
fib800_1=sky_lim[1][7][1:]

## At exposure time t=100 sec
fib100_2=sky_lim[2][0][1:]
fib150_2=sky_lim[2][1][1:]
fib200_2=sky_lim[2][2][1:]
fib300_2=sky_lim[2][3][1:]
fib400_2=sky_lim[2][4][1:]
fib500_2=sky_lim[2][5][1:]
fib600_2=sky_lim[2][6][1:]
fib800_2=sky_lim[2][7][1:]

## At exposure time t=300 sec
fib100_3=sky_lim[3][0][1:]
fib150_3=sky_lim[3][1][1:]
fib200_3=sky_lim[3][2][1:]
fib300_3=sky_lim[3][3][1:]
fib400_3=sky_lim[3][4][1:]
fib500_3=sky_lim[3][5][1:]
fib600_3=sky_lim[3][6][1:]
fib800_3=sky_lim[3][7][1:]

## At exposure time t=600 sec
fib100_4=sky_lim[4][0][1:]
fib150_4=sky_lim[4][1][1:]
fib200_4=sky_lim[4][2][1:]
fib300_4=sky_lim[4][3][1:]
fib400_4=sky_lim[4][4][1:]
fib500_4=sky_lim[4][5][1:]
fib600_4=sky_lim[4][6][1:]
fib800_4=sky_lim[4][7][1:]
```
